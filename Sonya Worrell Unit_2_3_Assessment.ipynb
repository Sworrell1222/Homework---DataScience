````python
# Answer 5 
import pandas as pd
import csv
import os
import matplotlib.pyplot as plt

#Import the bitcoin_cash_price.csv file into a Pandas dataframe called bitcoin

csv_path=os.path.join('bitcoin_cash_price.csv')
csv_path=os.path.join('dash_price.csv')
bitcoin = pd.read_csv('bitcoin_cash_price.csv')
#Import the dash_price.csv file into a Pandas dataframe called dash.
dash = pd.read_csv('dash_price.csv')
````

````python
#Merge the bitcoin and dash dataframes using the Date column as the identifier column. Name this new merged dataframe crypto.
crypto = pd.merge(bitcoin, dash, how="left", on ="Date")
crypto.head()
````

````python
#Create a new Delta_Volume column in the crypto dataframe that calculates the largest absolute difference between the bitcoin Volume versus dash Volume.
crypto["Delta_Volume"] = abs(crypto["Volume_x"]-crypto["Volume_y"])
crypto.head()
````
