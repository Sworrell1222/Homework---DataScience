```python
# Import dependencies
import pandas as pd
import os
```

```python
#Store file in variable path
coffeeshop_file = os.path.join('../','Resources', 'ratings_and_sentiments.csv') 
```

```python
# Read file into DataFrame
coffeeshop_df = pd.read_csv(coffeeshop_file)
coffeeshop_df
```

```python
# Find all reviews with a number rating greater than or equal to 4.
coffeeshop_df.loc[coffeeshop_df['num_rating']>=4, 'coffee_shop_name']
```

```python
# Find all the coffee shops with a overall sentiment greater than or equal to 4.
coffeeshop_df.loc[coffeeshop_df['overall_sent']>=4, 'coffee_shop_name']
```

```python
# Return the count of the above query.
coffeeshop_df.count()
```

```python
# Change the name of the review_text column to review
coffeeshop_df.rename(columns = {'review_text':'review'}, inplace=True)
```

```python
# Change the name of the bool_HIGH column to bool_high
coffeeshop_df.rename(columns = {'bool HIGH':'bool_high'}, inplace = True)
```

```python
# Save your changes to the file as a pipe delimited file.
coffeeshop_df.to_csv('rating_and_sentiments.csv', sep = '|')
```
